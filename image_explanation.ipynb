```python
import google.genai as genai
import google.genai.types as types
from dotenv import load_dotenv
import os

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

response = client.models.generate_content(
    model=os.getenv("GEMINI_MODEL", "gemini-2.0-flash"),
    contents=[
        types.Part(text="이 이미지에 대해 설명해주세요."),
        types.Part(
            file_data=types.FileData(
                file_uri="https://i.ytimg.com/vi/jAxSoBXN_5g/maxresdefault.jpg",
                mime_type="image/jpeg"
            )
        )
    ]
)

print(response.text)
```

```python
import base64

# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
    
image_path = r".\data\360_F_1537162917_C20wiVgwVtv1klja53NBIltckHJTAr5p.jpg"

# 이미지를 base64로 인코딩
base64_image = encode_image(image_path)

print(base64_image)
```

```python
response = client.models.generate_content(
    model=os.getenv("GEMINI_MODEL", "gemini-2.0-flash"),
    contents=[
        types.Part(text="이 이미지에 대해 설명해주세요."),
        types.Part(
            inline_data=types.Blob(
                mime_type="image/jpeg",
                data=base64_image  # base64 인코딩된 이미지 그대로 사용
            )
        )
    ]
)

print(response.text)
```

```python
seolleung_terrarosa_base64 = encode_image(r".\data\360_F_1076467513_QaPdOB3I0SqhkjN7HRvjFvhTPN8zsN5c.jpg")
local_stitch_terrarosa_base64 = encode_image(r".\data\360_F_1537162917_C20wiVgwVtv1klja53NBIltckHJTAr5p.jpg")

messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "두 카페의 차이점을 설명해주세요."},
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{seolleung_terrarosa_base64}",
                },
            },
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{local_stitch_terrarosa_base64}",
                },
            },
        ],
    }
]

response = client.models.generate_content(
    model=os.getenv("GEMINI_MODEL", "gemini-2.0-flash"),
    contents=[
        types.Part(text="두 사진의 차이점을 설명해주세요."),
        types.Part(
            inline_data=types.Blob(
                mime_type="image/jpeg",
                data=seolleung_terrarosa_base64
            )
        ),
        types.Part(
            inline_data=types.Blob(
                mime_type="image/jpeg",
                data=local_stitch_terrarosa_base64
            )
        )
    ]
)

print(response.text)
```

```python
oecd_rnd_2021_base64 = encode_image("./data/oecd_rnd_2021_large.png")
oecd_rnd_2022_base64 = encode_image("./data/oecd_rnd_2022.png")

response = client.models.generate_content(
    model=os.getenv("GEMINI_MODEL", "gemini-2.0-flash"),
    contents=[
        types.Part(text="첫번째는 2021년 데이터이고, 두번째는 2022년 데이터입니다. 이 데이터에 대해 설명해주세요. 어떤 변화가 있었나요? 한국 중심으로 설명해주세요."),
        types.Part(
            inline_data=types.Blob(
                mime_type="image/png",  # ← jpg가 아닌 png!
                data=oecd_rnd_2021_base64
            )
        ),
        types.Part(
            inline_data=types.Blob(
                mime_type="image/png",  # ← jpg가 아닌 png!
                data=oecd_rnd_2022_base64
            )
        )
    ]
)

print(response.text)
```
