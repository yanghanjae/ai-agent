```python
from langchain_community.document_loaders import PyPDFLoader

# PDF ьїїВЮ╝ВЮё ВЮйВќ┤Вёю ьЁЇВіцьіИ вЇ░ВЮ┤ьё░вЦ╝ ВХћВХюьЋЕвІѕвІц.
loader = PyPDFLoader('./data/OneNYC_2050_Strategic_Plan.pdf')
data_nyc = loader.load()
print(data_nyc)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# ьЁЇВіцьіИ вЇ░ВЮ┤ьё░вЦ╝ 1000Въљ вІеВюёвАю вѓўвѕЋвІѕвІц. overlapВЮђ 100ВъљвАю ВёцВаЋьЋЕвІѕвІц.
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
all_splits = text_splitter.split_documents(data_nyc)
```

```python
for i, split in enumerate(all_splits):
    print(f"Split {i+1}:------------------------------------\n")
    print(split)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
import os

load_dotenv()

embedding = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001",
    google_api_key=os.getenv("GEMINI_API_FREE_KEY") or os.getenv("GEMINI_API_KEY")
)

v = embedding.embed_query("вЅ┤ВџЋВЮў ВўеВІцЖ░ђВіц ВађЖ░љ ВаЋВ▒ЁВЮђ вГљВЋ╝?")
print(v)
print(len(v))
```

```python
from langchain_chroma import Chroma
import os

persist_directory = './chroma_store'	

# ВађВъЦвљю ьЂгвАювДѕ DBЖ░ђ ВЌєвІцвЕ┤ ВЃѕвАю вДївЊцЖИ░
if not os.path.exists(persist_directory):
    print("Creating new Chroma store")
    vectorstore = Chroma.from_documents(
        documents=all_splits,
        embedding=embedding,
        persist_directory=persist_directory
    )

else:
    print("Loading existing Chroma store")
    vectorstore = Chroma(		
        persist_directory=persist_directory, 
        embedding_function=embedding
    )
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

docs = retriever.invoke("ВёюВџИВІюВЮў ьЎўЖ▓й ВаЋВ▒ЁВЌљ вїђьЋ┤ ЖХЂЖИѕьЋ┤")

for d in docs:
    print(d)
    print('------')
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()

chat = ChatGoogleGenerativeAI(
    model=os.getenv("GEMINI_MODEL", "gemini-2.0-flash"),
    google_api_key=os.getenv("GEMINI_API_FREE_KEY") or os.getenv("GEMINI_API_KEY")
)

question_answering_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "ВѓгВџЕВъљВЮў ВДѕвгИВЌљ вїђьЋ┤ ВЋёвъў contextВЌљ ЖИ░в░ўьЋўВЌг вІхв│ђьЋўвЮ╝.:\n\n{context}",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

document_chain = create_stuff_documents_chain(chat, question_answering_prompt)
```

```python
from langchain_core.chat_history import InMemoryChatMessageHistory

# В▒ёьїЁ вЕћВІюВДђ ВађВъЦьЋа вЕћвфевдг Ж░ЮВ▓┤ ВЃЮВё▒
chat_history = InMemoryChatMessageHistory()

# ВѓгВџЕВъљ ВДѕвгИВЮё вЕћвфевдгВЌљ ВађВъЦ
chat_history.add_user_message("ВёюВџИВІюВЮў ВўеВІцЖ░ђВіц ВађЖ░љ ВаЋВ▒ЁВЌљ вїђьЋ┤ ВЋївацВцў.")

# вгИВёю Ж▓ђВЃЅьЋўЖ│а вІхв│ђ ВЃЮВё▒
answer = document_chain.invoke(
    {
        "messages": chat_history.messages,
        "context": docs,
    }
)

# ВЃЮВё▒вљю вІхв│ђ вЕћвфевдгВЌљ ВађВъЦ
chat_history.add_ai_message(answer)

print(answer)
```

```python
from langchain_core.output_parsers import StrOutputParser # вгИВъљВЌ┤ ВХюваЦ ьїїВёювЦ╝ вХѕвЪгВўхвІѕвІц.
```

```python
query_for_nyc = "вЅ┤ВџЋВЮђ?"

# query augmentation 
# ЖИ░ВА┤ вїђьЎћ вѓ┤ВџЕВЮё ьЎюВџЕьЋ┤ query_augmentation ВѕўьќЅ
query_augmentation_prompt = ChatPromptTemplate.from_messages(
    [
        MessagesPlaceholder(variable_name="messages"), # ЖИ░ВА┤ вїђьЎћ вѓ┤ВџЕ
        (
            "user",
            "ЖИ░ВА┤ВЮў вїђьЎћ вѓ┤ВџЕВЮё ьЎюВџЕьЋўВЌг ВѓгВџЕВъљВЮў ВЋёвъў ВДѕвгИВЮў ВЮўвЈёвЦ╝ ьїїВЋЁьЋўВЌг вфЁвБїьЋю ьЋю вгИВъЦВЮў ВДѕвгИВю╝вАю в│ђьЎўьЋўвЮ╝. вїђвфЁВѓгвѓў ВЮ┤, Вађ, ЖиИВЎђ Ж░ЎВЮђ ьЉюьўёВЮё вфЁьЎЋьЋю вфЁВѓгвАю ьЉюьўёьЋўвЮ╝. :\n\n{query}",
        ),
    ]
)
```

```python
query_augmentation_chain = query_augmentation_prompt | chat | StrOutputParser()
```

```python
augmented_query = query_augmentation_chain.invoke({
    "messages": chat_history.messages,
    "query": query_for_nyc  
})

print(augmented_query)
```

```python
# вгИВъљВЌ┤вАю в│ђьЎў
augmented_query = str(augmented_query)
print(type(augmented_query))
print(augmented_query)
```

```python
docs = retriever.invoke(augmented_query)

for d in docs:
    print(d)
    print('------')
```

```python
chat_history.add_user_message(query_for_nyc) # query_for_nycВЌљ "вЅ┤ВџЋВЮђ?" ВХћЖ░ђ

answer = document_chain.invoke(
    {
        "messages": chat_history.messages,
        "context": docs,
    }
)

# ВЃЮВё▒вљю вІхв│ђ вЕћвфевдгВЌљ ВађВъЦ
chat_history.add_ai_message(answer) 

print(answer)
```
