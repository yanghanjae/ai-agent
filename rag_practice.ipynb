```python
from langchain_community.document_loaders import PyPDFLoader

# PDF нҢҢмқјмқ„ мқҪм–ҙм„ң н…ҚмҠӨнҠё лҚ°мқҙн„°лҘј м¶”м¶ңн•©лӢҲлӢӨ.
loader = PyPDFLoader('./data/OneNYC_2050_Strategic_Plan.pdf')
data_nyc = loader.load()
print(data_nyc)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# н…ҚмҠӨнҠё лҚ°мқҙн„°лҘј 1000мһҗ лӢЁмң„лЎң лӮҳлҲ•лӢҲлӢӨ. overlapмқҖ 100мһҗлЎң м„Өм •н•©лӢҲлӢӨ.
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
all_splits = text_splitter.split_documents(data_nyc)
```

```python
for i, split in enumerate(all_splits):
    print(f"Split {i+1}:------------------------------------\n")
    print(split)
```

```python
loader_seoul = PyPDFLoader('./data/2040_seoul_plan.pdf')
data_seoul = loader_seoul.load()
seoul_splits = text_splitter.split_documents(data_seoul)
for i, split in enumerate(seoul_splits):
    print(f"Split {i+1}:------------------------------------")
    print(split)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
import os

load_dotenv()

embedding = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001",
    google_api_key=os.getenv("GEMINI_API_FREE_KEY") or os.getenv("GEMINI_API_KEY")
)

v = embedding.embed_query("лүҙмҡ•мқҳ мҳЁмӢӨк°ҖмҠӨ м Җк°җ м •мұ…мқҖ лӯҗм•ј?")
print(v)
print(len(v))
```

```python
from langchain_chroma import Chroma
import os

persist_directory = './chroma_store'	

# м ҖмһҘлҗң нҒ¬лЎңл§Ҳ DBк°Җ м—ҶлӢӨл©ҙ мғҲлЎң л§Ңл“Өкё°
if not os.path.exists(persist_directory):
    print("Creating new Chroma store")
    vectorstore = Chroma.from_documents(
        documents=all_splits,
        embedding=embedding,
        persist_directory=persist_directory
    )

else:
    print("Loading existing Chroma store")
    vectorstore = Chroma(		
        persist_directory=persist_directory, 
        embedding_function=embedding
    )
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

docs = retriever.invoke("м„ңмҡёмӢңмқҳ нҷҳкІҪ м •мұ…м—җ лҢҖн•ҙ к¶ҒкёҲн•ҙ")

for d in docs:
    print(d)
    print('------')
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()

chat = ChatGoogleGenerativeAI(
    model=os.getenv("GEMINI_MODEL", "gemini-2.0-flash"),
    google_api_key=os.getenv("GEMINI_API_FREE_KEY") or os.getenv("GEMINI_API_KEY")
)

question_answering_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "мӮ¬мҡ©мһҗмқҳ м§Ҳл¬ём—җ лҢҖн•ҙ м•„лһҳ contextм—җ кё°л°ҳн•ҳм—¬ лӢөліҖн•ҳлқј.:\n\n{context}",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

document_chain = create_stuff_documents_chain(chat, question_answering_prompt)
```

```python
from langchain_core.chat_history import InMemoryChatMessageHistory

# мұ„нҢ… л©”мӢңм§Җ м ҖмһҘн•  л©”лӘЁлҰ¬ к°қмІҙ мғқм„ұ
chat_history = InMemoryChatMessageHistory()

# мӮ¬мҡ©мһҗ м§Ҳл¬ёмқ„ л©”лӘЁлҰ¬м—җ м ҖмһҘ
chat_history.add_user_message("м„ңмҡёмӢңмқҳ мҳЁмӢӨк°ҖмҠӨ м Җк°җ м •мұ…м—җ лҢҖн•ҙ м•Ңл ӨмӨҳ.")

# л¬ём„ң кІҖмғүн•ҳкі  лӢөліҖ мғқм„ұ
answer = document_chain.invoke(
    {
        "messages": chat_history.messages,
        "context": docs,
    }
)

# мғқм„ұлҗң лӢөліҖ л©”лӘЁлҰ¬м—җ м ҖмһҘ
chat_history.add_ai_message(answer)

print(answer)
```

```python
from langchain_core.output_parsers import StrOutputParser # л¬ёмһҗм—ҙ м¶ңл Ҙ нҢҢм„ңлҘј л¶Ҳлҹ¬мҳөлӢҲлӢӨ.
```

```python
query_for_nyc = "лүҙмҡ•мқҖ?"

# query augmentation 
# кё°мЎҙ лҢҖнҷ” лӮҙмҡ©мқ„ нҷңмҡ©н•ҙ query_augmentation мҲҳн–ү
query_augmentation_prompt = ChatPromptTemplate.from_messages(
    [
        MessagesPlaceholder(variable_name="messages"), # кё°мЎҙ лҢҖнҷ” лӮҙмҡ©
        (
            "user",
            "кё°мЎҙмқҳ лҢҖнҷ” лӮҙмҡ©мқ„ нҷңмҡ©н•ҳм—¬ мӮ¬мҡ©мһҗмқҳ м•„лһҳ м§Ҳл¬ёмқҳ мқҳлҸ„лҘј нҢҢм•…н•ҳм—¬ лӘ…лЈҢн•ң н•ң л¬ёмһҘмқҳ м§Ҳл¬ёмңјлЎң ліҖнҷҳн•ҳлқј. лҢҖлӘ…мӮ¬лӮҳ мқҙ, м Җ, к·ёмҷҖ к°ҷмқҖ н‘ңнҳ„мқ„ лӘ…нҷ•н•ң лӘ…мӮ¬лЎң н‘ңнҳ„н•ҳлқј. :\n\n{query}",
        ),
    ]
)
```

```python
query_augmentation_chain = query_augmentation_prompt | chat | StrOutputParser()
```

```python
augmented_query = query_augmentation_chain.invoke({
    "messages": chat_history.messages,
    "query": query_for_nyc  
})

print(augmented_query)
```

```python
# л¬ёмһҗм—ҙлЎң ліҖнҷҳ
augmented_query = str(augmented_query)
print(type(augmented_query))
print(augmented_query)
```
